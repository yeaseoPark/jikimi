# 학교폭력신고수(경찰청)와 피해응답학생수(학교알리미) 상관관계 분석 
## 상관관계 corr(method='spearman')

```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
pd.set_option('display.max_rows', None)
```

```python
# 데이터 불어오기 
high_school = pd.read_csv("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\서울특별시_스케일링_고등학교지표.csv",index_col = 0)
high_school
```

```python
corr = high_school.groupby('시군구').mean().corr(method='spearman')
corr_ver2 = corr[['학교폭력신고수','피해 응답 학생 수']]
corr_ver2
```

# 유의미한 상관계수만 추출 

```python
meaning = pd.read_csv("C:/sooeun/DATAexam/ConvergenceProject-School_Violence/Data/서울특별시_시군구_스케일링_상관관계.csv")
meaning
```

```python
meaning.rename(columns = {'Unnamed: 0':'변수'},inplace=True)
meaning
```

```python
meaning_police = meaning[['변수','학교폭력신고수']]
```

```python
meaning_police_s = meaning_police[(meaning_police['학교폭력신고수'] <= -0.2)|(meaning_police['학교폭력신고수'] >= 0.2)]
meaning_police_s
```

```python
meaning_police_s_new = meaning_police_s.loc[[41,42,44,45,51,57,58,60],:]
meaning_police_s_new.rename(columns = {'학교폭력신고수':'상관관계'},inplace=True)
meaning_police_s_new
```

```python
meaning_school = meaning[['변수','피해 응답 학생 수']]
```

```python
meaning_school_s = meaning_school[(meaning_school['피해 응답 학생 수'] <= -0.2)|(meaning_school['피해 응답 학생 수'] >= 0.2)]
meaning_school_s
```

```python
meaning_school_s_new = meaning_school_s.loc[[47],:]
meaning_school_s_new.rename(columns = {'피해 응답 학생 수':'상관관계'},inplace=True)
meaning_school_s_new
```

```python
result = pd.concat([meaning_police_s_new,meaning_school_s_new], ignore_index=True)
result
```

```python
# 저장 후 주석처리 
result.to_csv("C:/sooeun/DATAexam/ConvergenceProject-School_Violence/Data/서울특별시_스케일링_상관관계_유의미한_변수.csv")
```
