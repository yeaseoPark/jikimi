```python
# 유해업소수 구하는 함수 
def close_Harmful(school_loc_data, Harmful_loc_data): 
    schoolloc, Harmfulloc, Harmful_cnt = [], [], []
    for col in tqdm(school_loc_data.itertuples()): 
        schoolloc = [col.위도, col.경도]
        distance = []
        for col2 in Harmful_loc_data.itertuples(): 
            Harmfulloc = [col2.위도, col2.경도]
            distance.append(haversine(schoolloc, Harmfulloc, unit = 'm'))
            
        distance = [d for d in distance if d <= 500] 
        Harmful_cnt.append(len(distance))
    return Harmful_cnt
```
