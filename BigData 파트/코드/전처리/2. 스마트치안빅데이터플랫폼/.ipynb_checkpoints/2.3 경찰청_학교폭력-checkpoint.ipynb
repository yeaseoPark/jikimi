```python
# 경찰청_학교폭력데이터 불러오기 
# index_col = 0 : 첫 번째 컬럼을 index로 사용하도록 지정하여 로드 
school_violence_data_seoul = pd.read_csv("C:/sooeun/DATAexam/ConvergenceProject-School_Violence/Data/경찰청_2021_서울_학교폭력데이터.csv",index_col = 0)
school_violence_data_seoul
```

```python
# 접수년월 2021로 변경 
school_violence_data_seoul = school_violence_data_seoul.replace({'접수년월' : 202101}, 2021) 
school_violence_data_seoul = school_violence_data_seoul.replace({'접수년월' : 202102}, 2021) 
school_violence_data_seoul = school_violence_data_seoul.replace({'접수년월' : 202103}, 2021) 
school_violence_data_seoul = school_violence_data_seoul.replace({'접수년월' : 202104}, 2021) 
school_violence_data_seoul = school_violence_data_seoul.replace({'접수년월' : 202105}, 2021) 
school_violence_data_seoul = school_violence_data_seoul.replace({'접수년월' : 202106}, 2021) 
school_violence_data_seoul = school_violence_data_seoul.replace({'접수년월' : 202107}, 2021) 
school_violence_data_seoul = school_violence_data_seoul.replace({'접수년월' : 202108}, 2021) 
school_violence_data_seoul = school_violence_data_seoul.replace({'접수년월' : 202109}, 2021) 
school_violence_data_seoul = school_violence_data_seoul.replace({'접수년월' : 202110}, 2021) 
school_violence_data_seoul = school_violence_data_seoul.replace({'접수년월' : 202111}, 2021) 
school_violence_data_seoul = school_violence_data_seoul.replace({'접수년월' : 202112}, 2021) 
```

```python
# 접수년월 -> 접수년도
school_violence_data_seoul.rename(columns = {'접수년월':'접수년도'},inplace=True)
school_violence_data_seoul
```

```python
# 경찰서명 서울 자르고 시군구 열 추가하기 
school_violence_data_seoul["시군구"]=school_violence_data_seoul["경찰서명"].str[2:5]
```

```python
school_violence_data_seoul
```

```python
# 시군구 수정하기 1 - 구 붙이기 
school_violence_data_seoul['시군구'] = school_violence_data_seoul['시군구'] + "구"
```

```python
# 시군구 수정하기1 - 경찰청 시군구 위치로 시군구 매핑 
school_violence_data_seoul = school_violence_data_seoul.replace({'시군구' : '수서구'}, '강남구')
school_violence_data_seoul = school_violence_data_seoul.replace({'시군구' : '중부구'}, '중구')
school_violence_data_seoul = school_violence_data_seoul.replace({'시군구' : '종암구'}, '성북구')
school_violence_data_seoul = school_violence_data_seoul.replace({'시군구' : '서부구'}, '은평구')
school_violence_data_seoul = school_violence_data_seoul.replace({'시군구' : '방배구'}, '서초구')
school_violence_data_seoul = school_violence_data_seoul.replace({'시군구' : '혜화구'}, '종로구')
school_violence_data_seoul = school_violence_data_seoul.replace({'시군구' : '남대문구'}, '중구')
school_violence_data_seoul = school_violence_data_seoul.replace({'시군구' : '청구'}, '종로구')
```
