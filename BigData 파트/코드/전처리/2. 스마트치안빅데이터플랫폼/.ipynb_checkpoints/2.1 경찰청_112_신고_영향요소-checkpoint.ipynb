```python
# 112 신고 요소 불러오기
# index_col = 0 : 첫 번째 컬럼을 index로 사용하도록 지정하여 로드 
data_01 = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/112_신고_영향요소/FSN_08_20210131_C_001.csv",header = None)
data_02 = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/112_신고_영향요소/FSN_08_20210228_C_001.csv",header = None)
data_03 = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/112_신고_영향요소/FSN_08_20210331_C_001.csv",header = None)
data_04 = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/112_신고_영향요소/FSN_08_20210430_C_001.csv",header = None)
data_05 = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/112_신고_영향요소/FSN_08_20210531_C_001.csv",header = None)
data_06 = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/112_신고_영향요소/FSN_08_20210630_C_001.csv",header = None)
data_07 = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/112_신고_영향요소/FSN_08_20210731_C_001.csv",header = None)
data_08 = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/112_신고_영향요소/FSN_08_20210831_C_001.csv",header = None)
data_09 = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/112_신고_영향요소/FSN_08_20210930_C_001.csv",header = None)
data_10 = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/112_신고_영향요소/FSN_08_20211031_C_001.csv",header = None)
data_11 = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/112_신고_영향요소/FSN_08_20211130_C_001.csv",header = None)
data_12 = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/112_신고_영향요소/FSN_08_20211231_C_001.csv",header = None)
```

```python
# 데이터 합치기
data = pd.concat([data_01, data_02, data_03, data_04, data_05, data_06, data_07,
                  data_08, data_09, data_10, data_11, data_12], axis=0)
data
```

```python
# 컬럼명 변경 
data.columns = data_headers[' 컬럼한글명']
data
```

```python
# 인덱스 재설정
data.reset_index(drop=True, inplace=True)
data
```

```python
seoul_data = data[data['경찰서명'].str.contains('서울')]
seoul_data
```

```python
seoul_data.reset_index(drop=True, inplace=True)
seoul_data
```

```python
# 경찰서명 서울 자르고 시군구 열 추가하기 
seoul_data["시군구"]=seoul_data["경찰서명"].str[2:5]
seoul_data
```

```python
# 시군구 수정하기 1 - 구 붙이기 
seoul_data['시군구'] = seoul_data['시군구'] + "구"
seoul_data
```

```python
 # 시군구 수정하기1 - 경찰청 시군구 위치로 시군구 매핑 
seoul_data = seoul_data.replace({'시군구' : '수서구'}, '강남구')
seoul_data = seoul_data.replace({'시군구' : '중부구'}, '중구')
seoul_data = seoul_data.replace({'시군구' : '종암구'}, '성북구')
seoul_data = seoul_data.replace({'시군구' : '서부구'}, '은평구')
seoul_data = seoul_data.replace({'시군구' : '방배구'}, '서초구')
seoul_data = seoul_data.replace({'시군구' : '혜화구'}, '종로구')
seoul_data = seoul_data.replace({'시군구' : '남대문구'}, '중구')
seoul_data = seoul_data.replace({'시군구' : '청구'}, '종로구')
seoul_data
```

```python
seoul_data.groupby('시군구').sum().columns
```

```python
# 구별 평균 계산
seoul_data_fin = seoul_data.groupby('시군구').mean()
seoul_data_fin
```

```python
seoul_data_fin.columns
```

```python
# 지수표현 방지
pd.options.display.float_format = '{:.5f}'.format
```

```python
seoul_data_fin = seoul_data_fin[['신고건수전체', '사건종별가정폭력', '사건종별데이트폭력', '월평균소득금액', '총상주인구수']]
seoul_data_fin
```

```python
school_violence = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/경찰청_2021_서울_시군구처리_학교폭력데이터.csv")
school_violence
```

```python
school_violence.drop('Unnamed: 0', axis=1, inplace=True)
school_violence.index = school_violence['시군구']
school_violence.drop('시군구', axis=1, inplace=True)
school_violence
```

```python
seoul_data_fin['학교폭력 신고 수'] = school_violence['신고건수전체']
seoul_data_fin
```
