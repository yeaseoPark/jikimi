# 시군구 별 학교폭력피해장소

```python
import pandas as pd
import numpy as np
```

```python
place_df = pd.read_csv("C:/sooeun/DATAexam/ConvergenceProject-School_Violence/Data/학교알리미_2021_서울_학교폭력피해장소.csv")
place_df
```

```python
place_df2 = place_df.groupby(["시군구"]).agg({"교실안":"sum","특별실,방과후 교실 등":"sum","복도,계단":"sum","화장실":"sum","운동장,체육관,강당 등":"sum","급식실,매점 등":"sum",
                                          "기숙사":"sum","학원이나학원 근처":"sum","집이나집 근처":"sum","PC방,노래방,오락실 등":"sum","공원, 놀이터,골목, 공터,뒷산 등":"sum","학교 밖체험활동 장소":"sum",
                                          "사이버공간":"sum","기타":"sum"}).reset_index()
place_df2
```

```python
place_df2.rename(columns = {'기타':'피해장소기타'},inplace=True)
place_df2
```

```python
place_df2['피해장소합계']=place_df2.sum(axis=1)
place_df2
```
