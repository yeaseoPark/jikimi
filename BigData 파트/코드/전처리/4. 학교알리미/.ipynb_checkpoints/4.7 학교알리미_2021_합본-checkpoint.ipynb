```python
# 1. 나이스_학교정보 데이터 불러오기 
school = pd.read_csv("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\나이스_서울_전체_위경도_학교기본정보_Ver2.csv",index_col = 0)
school
```

```python
# 2. 학교폭력피해장소 데이터 불러오기 
place = pd.read_csv("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\학교알리미_2021_서울_학교폭력피해장소.csv")
place.drop(['시군구'], axis = 1,inplace=True)
place.drop(['구분'], axis = 1,inplace=True)
```

```python
# 3. 학교폭력피해유형 데이터 불러오기 
typedf = pd.read_csv("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\학교알리미_2021_서울_학교폭력피해유형.csv")
typedf.drop(['시군구'], axis = 1,inplace=True)
typedf.drop(['구분'], axis = 1,inplace=True)
```

```python
# 4. 학교폭력피해시간 데이터 불러오기 
time = pd.read_csv("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\학교알리미_2021_서울_학교폭력피해시간.csv")
time.drop(['시군구'], axis = 1,inplace=True)
time.drop(['구분'], axis = 1,inplace=True)
```

```python
# 5. 학교폭력교육시간 데이터 불러오기 
edu = pd.read_csv("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\학교알리미_2021_서울_학교폭력교육시간.csv")
edu.drop(['시군구'], axis = 1,inplace=True)
```

```python
# 6. 학교폭력실태조사 참여현황 데이터 불러오기 
answer = pd.read_csv("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\학교알리미_2021_서울_학교폭력실태조사_참여현황.csv")
answer.drop(['시군구'], axis = 1,inplace=True)
answer.drop(['구분'], axis = 1,inplace=True)
```

```python
# 7. 학교폭력피해신고현황 데이터 불러오기 
noti = pd.read_csv("C:/sooeun/DATAexam/ConvergenceProject-School_Violence/Data/학교알리미_2021_서울_학교폭력피해신고현황.csv")
noti.drop(['시군구'], axis = 1,inplace=True)
noti.drop(['구분'], axis = 1,inplace=True)
```

```python
# 8. 500m_cctv수_경찰서수
count1 = pd.read_csv("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\학교500m_경찰서_개수_Ver2.csv")
count1 = count1[['학교명','500m_inner_police_dept']]
```

```python
# 9. 500m_유흥업소수
count2 = pd.read_csv("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\학교500이내_유해업소수_Ver2.csv")
count2.drop(['Unnamed: 0'], axis = 1,inplace=True)
count2 = count2[['학교명','500m_inner_Harmful_dept']]
```

```python
# 10. 500m_cctv수
count3 = pd.read_csv("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\학교500m_cctv_개수_Ver2.csv")
count3 = count3[['학교명','500m_inner_cctv']]
```

```python
data1 = pd.merge(school, place, on=['학교명'], how='outer', sort=True)
data2 = pd.merge(data1, typedf, on=['학교명'], how='outer', sort=True)
data3 = pd.merge(data2, time, on=['학교명'], how='outer', sort=True)
data4 = pd.merge(data3, edu, on=['학교명'], how='outer', sort=True)
data5 = pd.merge(data4, answer, on=['학교명'], how='outer', sort=True)
data6 = pd.merge(data5, noti, on=['학교명'], how='outer', sort=True)
data7 = pd.merge(data6, count1, on=['학교명'], how='outer', sort=True)
data8 = pd.merge(data7, count2, on=['학교명'], how='outer', sort=True)
data9 = pd.merge(data8, count3, on=['학교명'], how='outer', sort=True)
data9
```
