# 시군구 별 학교폭력피해유형

```python
import pandas as pd
import numpy as np
```

```python
type_df2 = pd.read_csv("C:/sooeun/DATAexam/ConvergenceProject-School_Violence/Data/학교알리미_2021_서울_학교폭력피해유형.csv")
type_df2
```

```python
type_df3 = type_df2.groupby(["시군구"]).agg({"강요":"sum","금품갈취":"sum","신체폭력":"sum","사이버폭력":"sum","집단따돌림":"sum","기타":"sum"}).reset_index()
type_df3
```

```python
type_df3.rename(columns = {'기타':'피해유형기타'},inplace=True)
type_df3
```

```python
type_df3['피해유형합계']=type_df3.sum(axis=1)
type_df3
```
