# 시군구 별 학교폭력피해시간 

```python
import pandas as pd
import numpy as np
```

```python
da_time_df = pd.read_csv("C:/sooeun/DATAexam/ConvergenceProject-School_Violence/Data/학교알리미_2021_서울_학교폭력피해시간.csv")
da_time_df
```

```python
da_time_df = da_time_df.groupby(["시군구"]).agg({"등교시간":"sum","수업시간":"sum","쉬는시간":"sum","점심시간":"sum","학교 안방과후 시간":"sum",
                                          "하교시간":"sum","학교 밖체험활동 시간":"sum","학교 일과가끝난 후":"sum","기타":"sum"}).reset_index()
da_time_df
```

```python
da_time_df.rename(columns = {'기타':'피해시간기타'},inplace=True)
da_time_df
```

```python
da_time_df['피해시간합계']=da_time_df.sum(axis=1)
da_time_df
```
