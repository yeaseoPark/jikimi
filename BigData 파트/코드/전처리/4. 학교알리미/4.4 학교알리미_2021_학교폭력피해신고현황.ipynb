```python
# 은일정보산업고등학교->유한공업고등학교
# 덕수고등학교(일반고) 삭제 
# 송파구 가락고등학교 

#나이스_학교기본정보 / 학교알리미_학교폭력실태조사 학교명 비교 
# place = pd.read_csv("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\학교알리미_2021_서울_학교폭력피해장소.csv")

#데이터 불러오기 
edu = pd.read_csv("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\학교알리미_2021_서울_학교폭력교육시간.csv")

school_info_name = pd.DataFrame(edu['학교명'])
high_school_name = pd.DataFrame(noti['학교명'])

def GetDifferences(school_info_name, high_school_name):
  df = pd.concat([school_info_name, high_school_name]).reset_index(drop=True)
  idx = [diff[0] for diff in df.groupby(list(df.columns)).groups.values() if len(diff) == 1]
  return df.reindex(idx)

GetDifferences(school_info_name, high_school_name)
```
