# 소득인식수준 데이터(서울열린데이터광장, 서울특별시)

```python
import pandas as pd
import numpy as np
```

```python
income_recognition_level = pd.read_csv("D:/MultiCampus/DATAexam/융복합/폭력분석/data/소득/소득인식수준_20230117154458.csv")
income_recognition_level
```

```python
income_recognition_level[income_recognition_level['구분별(1)'] == '지역소분류']
```

```python
# 최신 데이터만 slicing
income_recognition_level = income_recognition_level[['구분별(2)', '수준별(2)', '2017']]
income_recognition_level
```

```python
income_recognition_level = income_recognition_level[105:len(income_recognition_level)]
income_recognition_level
```

```python
income_recognition_level.reset_index(drop=True, inplace=True)
income_recognition_level
```

```python
income_recognition_level.rename(columns={'구분별(2)':'시군구', '수준별(2)': '소득인식', '2017': '비율'}, inplace=True)
income_recognition_level
```

```python
income_pt = pd.pivot_table(income_recognition_level,
                           index = income_recognition_level['시군구'],
                           columns = income_recognition_level['소득인식'])
income_pt
```

```python
income_pt.columns
```

```python
income_pt.columns = income_pt.columns.get_level_values(1)
income_pt
```

```python
income_pt.columns.name=''
```

```python
income_pt = income_pt[['저소득', '중간', '고소득']]
income_pt
```

```python
income_pt.to_csv('D:/MultiCampus/DATAexam/융복합/폭력분석/data/구별_소득_분석.csv', encoding='utf-8-sig')
```
