```python
#행정구역코드 매핑정보
code = pd.read_excel("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\자치구코드.xlsx")
code
```

```python
code.columns
```

```python
code.drop('Unnamed: 3', axis=1, inplace=True)
code.drop('Unnamed: 4', axis=1, inplace=True)
code.drop('RESD_DO_NM', axis=1, inplace=True)
code
```

```python
code.rename(columns = {'RESD_CD':'행정구역코드','RESC_CT_NM':'시군구명'},inplace=True)
code
```

```python
#불필요한 행 삭제
code.drop(labels=range(25, 91), axis=0, inplace=True)
code
```

```python
code['행정구역코드'] = code['행정구역코드'].astype(np.int64)
code.info()
```

```python
gu = []
for i in range(len(floating_population_sliced)):
    floating_population_code = floating_population_sliced.자치구코드.iloc[i]
    code_index = code.index[code.행정구역코드 == floating_population_code][0]
    gu.append(code.시군구명.iloc[code_index])
```

```python
floating_population_mapped = floating_population_sliced
floating_population_mapped['행정동코드'] = gu
floating_population_mapped
```

```python
floating_population_mapped.drop(['자치구코드'],axis=1, inplace=True)
floating_population_mapped
```

```python
floating_population_mapped=floating_population_mapped.groupby(['행정동코드','기준일ID']).mean()
```

```python
floating_population_mapped.drop(['시간대구분'],axis=1, inplace=True)
floating_population_mapped
```

```python
floating_population_mapped['생활인구수']=floating_population_mapped['남자15세부터19세생활인구수']+floating_population_mapped['여자15세부터19세생활인구수']
floating_population_mapped
floating_population_mapped.drop(['남자15세부터19세생활인구수'],axis=1, inplace=True)
floating_population_mapped.drop(['여자15세부터19세생활인구수'],axis=1, inplace=True)
floating_population_mapped

```

```python
floating_population_mapped=floating_population_mapped.groupby(['행정동코드']).mean().reset_index()
floating_population_mapped
```

```python
floating_population_mapped.rename(columns = {'행정동코드':'시군구'},inplace=True)
floating_population_mapped
```
