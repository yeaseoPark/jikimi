```python
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기 
family = pd.read_csv("C:\sooeun\DATAexam\ConvergenceProject-School_Violence\Data\보통가구(동별)_20230125163339.csv",index_col = 0)
family 
```

```python
family_df = family[family['구분별(2)'].str.contains('소계', na = False)]
```

```python
family_df[['동별(2)','2010']]
```

```python
# 인덱스 수정 
family_df.reset_index(drop=False, inplace=True)

# # 'index' 컬럼 drop
# family_df = family_df.drop(['index'],axis='columns')
# family_df
```

```python
family_df['가구수'] = family_df['2010'].astype('int64')
```

```python
family_df.dtypes
```

```python
family_df = family_df.groupby(["동별(2)"]).agg({"가구수":"sum"}).reset_index()
family_df
```

```python
family_df.rename(columns = {'동별(2)':'시군구'},inplace=True)
family_df
```

```python
family_df = family_df.drop(index=17,axis=0)
family_df.reset_index(drop=False, inplace=True)
```

```python
family_df = family_df.drop(['index'],axis='columns')
family_df
```
